### then we found it is not related to car_name so we tried to replace them by their real values after searching it up in google

```python
c=0
arr=[75,140,53.5,120,81,82] ## this array has real horse power values of the cars with-10000 horsepower in order according to index
for i in df[df['horsepower']<0].index:
    df.iloc[i,2]=arr[c]
    c=c+1  
```
